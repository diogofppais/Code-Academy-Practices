````python
import pandas as pd
pd.set_option('display.max_colwidth', None)

#Importing the CSV file
jeopardy = pd.read_csv('jeopardy.csv')

#Renaming columns
jeopardy = jeopardy.rename(columns = {" Air Date": "Air Date", " Round" : "Round", " Category": "Category", " Value": "Value", " Question":"Question", " Answer": "Answer"})

print(jeopardy.columns)
````

2. Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list `["King", "England"]` was passed to our function, the function returned a DataFrame of 49 rows. Every row had the strings `"King"` and `"England"` somewhere in its `" Question"`.

   Test your function by printing out the column containing the question of each row of the dataset.

````python
# Write a function to filter the dataset based on a list of words
def filter_questions_by_words(data, word_list):
    condition = data['Question'].apply(lambda question: all(word in question for word in word_list))
    filtered_data = data[condition]
    return filtered_data

# List of words to filter by
filter_words = ["King", "England"]

# Apply the filtering function and get the filtered DataFrame
filtered_df = filter_questions_by_words(jeopardy, filter_words)

# Print the "Question" column for each row in the filtered DataFrame
for question in filtered_df['Question']:
    print(question)
````

3. Test your original function with a few different sets of words to try to find some ways your function breaks. Edit your function so it is more robust.

   For example, think about capitalization. We probably want to find questions that contain the word `"King"` or `"king"`.
   
   You may also want to check to make sure you don't find rows that contain substrings of your given words. For example, our function found a question that didn't contain the word `"king"`, however it did contain the word `"viking"` &mdash; it found the `"king"` inside `"viking"`. Note that this also comes with some drawbacks &mdash; you would no longer find questions that contained words like `"England's"`.

````python
# Enhanced function to filter the dataset based on a list of words
def filter_questions_by_words(data, word_list):
    filtered_data = pd.DataFrame()
    for word in word_list:
        condition = data['Question'].str.contains(rf'\b{word}\b', case=False)
        filtered_data = pd.concat([filtered_data, data[condition]])
    return filtered_data

# List of words to filter by
filter_words = ["King", "England"]

# Apply the enhanced filtering function and get the filtered DataFrame
filtered_df = filter_questions_by_words(jeopardy, filter_words)

# Print the "Question" column for each row in the filtered DataFrame
for question in filtered_df['Question']:
    print(question)
````

4. We may want to eventually compute aggregate statistics, like `.mean()` on the `" Value"` column. But right now, the values in that column are strings. Convert the`" Value"` column to floats. If you'd like to, you can create a new column with float values.

   Now that you can filter the dataset of question, use your new column that contains the float values of each question to find the "difficulty" of certain topics. For example, what is the average value of questions that contain the word `"King"`?
   
   Make sure to use the dataset that contains the float values as the dataset you use in your filtering function.

````python
# Adding a new column. If the value of the float column is not "None", then we cut off the first character (which is a dollar sign), and replace all commas with nothing, and then cast that value to a float. If the answer was "None", then we just enter a 0.
jeopardy["Float Value"] = jeopardy["Value"].apply(lambda x: float(x[1:].replace(',','')) if x != "None" else 0)

# Filtering the dataset and finding the average value of those questions
filtered = filter_questions_by_words(jeopardy, ['King'])
print(filtered["Float Value"].mean())
````
